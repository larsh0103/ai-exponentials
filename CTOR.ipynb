```python
import matplotlib.pyplot as plt
import numpy as np

# values of neural rendering frames per second
papers = {"2020": "https://arxiv.org/pdf/2003.08934.pdf", "2021": "https://arxiv.org/pdf/2103.03231.pdf", "2022": "https://arxiv.org/pdf/2208.00277.pdf"}
papers_names = {"2020": "Nerf", "2021": "DoNerf", "2022": "MobileNerf"}
ms_frame = {"2020": 1 / 3.2, "2021": 1 / 0.067, "2022": 1 / (1 / 744.91)}

# plot graph
plt.bar(range(len(ms_frame)), list(ms_frame.values()), align='center')
plt.xticks(range(len(ms_frame)), [papers_names[key] for key in ms_frame.keys()])
plt.title("Neural Rendering Frames Per Second")
# add labels
plt.xlabel("Paper")
plt.ylabel("Frames Per Second")

plt.show()

# fit to a exponential curve
x = np.array([int(k) for k in ms_frame.keys()])
y = np.array(list(ms_frame.values()))

# fit to a exponential curve
z = np.polyfit(x, np.log(y), 1)
f = np.poly1d(z)

# calculate new x's and y's
x_new = np.linspace(x[0], x[-1], 50)
y_new = np.exp(f(x_new))

# plot graph
plt.plot(x, y, 'o', x_new, y_new)
plt.xlim([2020, 2023])
# plt.xticks(x, [papers_names[str(xi)] for xi in x])  # paper name for x-axis labels
# add title
plt.title("Neural Rendering Frames Per Second")
# add labels
plt.xlabel("Year")
plt.ylabel("Frames Per Second")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# define key data points
ms_frame = {"August-2022": 1/5.6, "Dec-2022":(1/5.6)*20,"April-2023": 30}
papers = {"August-2022":"https://arxiv.org/pdf/2112.10752.pdf","Dec-2022":"https://arxiv.org/pdf/2210.03142.pdf","April-2023":"no paper yet"}
paper_names = {"August-2022":"Stable Diffusion","Dec-2022":"On Distillation of Guided Diffusion Models","April-2023":"Rumored 30 fps stable diffusion"}

# plot bar graph
plt.bar(range(len(ms_frame)), list(ms_frame.values()), align='center')
plt.xticks(range(len(ms_frame)), list(ms_frame.keys()))
plt.title("Image generation - Frames per Second")
plt.xlabel("Date")
plt.ylabel("Frames per Second")
plt.show()

# fit to an exponential curve
x_data = np.array([0, 3, 8])  # index-based representation for month intervals
y_data = np.array(list(ms_frame.values()))

# create a function that represents an exponential model
def exp_func(x, a, b, c):
    return a * np.exp(b * x) + c

# perform curve fitting to find the optimal parameters
from scipy.optimize import curve_fit
params, _ = curve_fit(exp_func, x_data, y_data)

# calculate new x's and y's
x_new = np.linspace(x_data[0], x_data[-1], 50)
y_new = exp_func(x_new, *params)

# plot curve
plt.plot(x_data, y_data, 'o', label='Original data')
plt.plot(x_new, y_new, label='Fitted curve')
plt.xticks(x_data, list(ms_frame.keys()))

plt.title("Image generation - Frames per Second (Exp Curve)")
plt.xlabel("Date")
plt.ylabel("Frames per Second")
plt.legend()
plt.show()
```

```python
### Now we make a plot about langauge models, focusing on gpt like models. 
# Key datapoints, novemeber 2022 chatgpt is released, and it has 175 billion parameters and blows the competition out of the water. It reportedly cost 10-20 million dollars to train.
# In march 2023 Alpaca is released by standford, it is only 7 billion parameters, runs on a phone, and matches the performance of chatgpt. It is open source and free to use.

import matplotlib.pyplot as plt
import numpy as np

# define key data points
language_models = {"November-2022 Davinci-003 ": 175, "March-2023 Alpaca": 7}
papers = {"November-2022 Davinci-003":"https://arxiv.org/pdf/2203.02155.pdf","March-2023 Alpaca":"https://crfm.stanford.edu/2023/03/13/alpaca.html"}

# plot bar graph
plt.bar(range(len(language_models)), list(language_models.values()), align='center')
plt.xticks(range(len(language_models)), list(language_models.keys()))
plt.title("Language Models - Number of Parameters")
plt.xlabel("Release Date")
plt.ylabel("Number of Parameters (billion)")
plt.show()

# fit to an exponential curve
x = np.array([int(k.split("-")[1].split(" ")[0]) for k in language_models.keys()])
y = np.array(list(language_models.values()))
z = np.polyfit(x, np.log(y), 1)
f = np.poly1d(z)

# calculate new x's and y's
x_new = np.linspace(x[0], x[-1], 50)
y_new = np.exp(f(x_new))

# plot curve
plt.plot(x, y, 'o', x_new, y_new)
plt.xlim([2022, 2023])
plt.title("Language Models - Number of Parameters (Exp Curve)")
plt.xlabel("Year")
plt.ylabel("Number of Parameters (billion)")
plt.show()
```
